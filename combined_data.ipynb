```python
import pandas as pd
```

```python
alignment_successful = pd.read_csv("alignment_successful.csv")
alignment_successful = alignment_successful.drop("Unnamed: 0", axis = 1).drop_duplicates()
alignment_successful.columns
```

```python
big_sentiment = pd.read_csv("Big Sentiment.csv")
big_sentiment = big_sentiment.drop("Unnamed: 0", axis = 1).drop_duplicates()
big_sentiment.columns
```

```python
cadence_match = pd.read_csv("Cadence Match.csv")
cadence_match = cadence_match.drop(["Unnamed: 0", "match_length"], axis = 1).drop_duplicates()
cadence_match.columns
```

```python
mudac_clean = pd.read_excel("MUDAC_clean.xlsx").drop_duplicates()
mudac_clean.columns
```

```python
training_cleaned = pd.read_csv("Training_cleaned_Anh.csv")
training_cleaned = training_cleaned.drop(["Unnamed: 0", 'Big Home = Little Home', 'Same County',
       'Same State', 'Same Zip'], axis = 1).drop_duplicates()
training_cleaned.columns
```

```python
cleaned_data_inti = pd.read_excel("Data\clean_data_inti.xlsx")
cleaned_data_inti = cleaned_data_inti.drop(["Little ID", "Big ID", 'ReEnroll_NA.y', 'ReEnroll_Yes.y', 'ReEnroll_No.y'], axis = 1)
cleaned_data_inti.columns
```

```python
cleaned_data_inti_match_note_avg = cleaned_data_inti.groupby("Match ID 18Char")[["Days Match Activation to Note Completion"]].mean().reset_index()
cleaned_data_inti = cleaned_data_inti.drop(["Days Match Activation to Note Completion", 'is_not_NA_Days Match Activation to Note Completion'], axis = 1).drop_duplicates()
cleaned_data_inti = cleaned_data_inti.merge(cleaned_data_inti_match_note_avg, on = "Match ID 18Char", how = "inner")
cleaned_data_inti
```

```python
df = training_cleaned.merge(alignment_successful, how = "inner", on = "Match ID 18Char")
print(df.shape)
df = df.merge(big_sentiment.drop("match_length", axis = 1), how = "left", left_on = "Match ID 18Char", right_on = "match_id_18char")
print(df.shape)
df = df.merge(cadence_match, how = "left", left_on = "Match ID 18Char", right_on = "matchid")
print(df.shape)
df = df.merge(mudac_clean.drop("Match Length", axis = 1), how = "left", on = "Match ID 18Char")
print(df.shape)
df = df.merge(cleaned_data_inti, on = "Match ID 18Char", how = "inner")
print(df.shape)
df.columns
```
